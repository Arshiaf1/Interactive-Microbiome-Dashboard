# 01: Data Exploration

This notebook performs basic exploratory data analysis (EDA) on the processed microbiome dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Metadata
df_meta = pd.read_csv('../data/processed/metadata.csv', index_col=0)
display(df_meta.head())
```

```python
# Summarize body sites
sns.countplot(data=df_meta, x='body-site')
plt.title('Samples per Body Site')
plt.show()
```
