# 02: Microbiome Analysis

This notebook covers diversity analysis and taxonomic breakdown.

```python
import pandas as pd
import numpy as np
import plotly.express as px

df_alpha = pd.read_csv('../data/processed/alpha_diversity.csv', index_col=0)
df_meta = pd.read_csv('../data/processed/metadata.csv', index_col=0)
df_merged = df_alpha.join(df_meta)

fig = px.box(df_merged, x='body-site', y='shannon', color='subject', title='Shannon Diversity by Body Site')
fig.show()
```
